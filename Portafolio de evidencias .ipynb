<h4 Style = "color: #A0522D"> 2.3 Imputacion simple de valores numéricos </h4>


<ol>
    <li>No todos los datos se van a imputar </li> 
    <li>Colocar observaciones cuando faltan datos</li> 
    <li>Imputar valores por la mediana aritmetica si es que el dato no es sensible
    si el dato es muy sensible, debe de llenarse</li> 
    <li>Sustituir perdidas normalizadas, hay que usar el promedio.
    Teorema Central de Limite : si se tiene un conjunto de valores de 1 a n, cada vez que se suma un 1+n con media aritmetica los datos se distribuyen</li> 
   
    
</ol>


```python
promedionormalizadas= dfAutos["normalized"].astype("float").mean(axis=0)
dfAutos["normalized"].replace(np.nan, promedionormalizadas, inplace=True)
dfAutos.head(5)
```
